We will use phi-2 model here and try do some basic tasks with it

```python
# libraries
from transformers import pipeline
from transformers import AutoModelForCausalLM, AutoTokenizer
```

```python
repo_name = "microsoft/phi-2"
```

```python
pipe = pipeline(task="text-generation",model=repo_name)
```

```python
prompt= "I am over using social media, can it be a problem?"
```

```python
res = pipe(prompt,max_new_tokens=100)
```
